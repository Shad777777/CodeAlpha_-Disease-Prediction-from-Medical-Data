```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
from matplotlib.figure import Figure
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline

from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('//content//heart.csv')
df
```

```python
df['Cholesterol'] = df['Cholesterol'].replace(0, df['Cholesterol'].median())
df_no_ecg = df.drop(columns=['ST_Slope','RestingECG','Oldpeak'])
df_no_ecg['Cholesterol'] = df_no_ecg['Cholesterol'].replace(0,df_no_ecg['Cholesterol'].median())
```

```python
df_no_ecg
```

```python
df_no_ecg.info()
```

Setting up categrory encoding for the catrgorical data

```python
import category_encoders as ce
```

```python
encoder = ce.OneHotEncoder(use_cat_names=True)
```

```python
y = df_no_ecg['HeartDisease']
X = df_no_ecg.drop(columns=['HeartDisease'], axis='columns')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
```

```python
print('Train',X_train.shape, y_train.shape)
print('Test',X_test.shape, y_test.shape)
```

```python
encoder.fit(X_train)
X_train = encoder.transform(X_train)
X_test = encoder.transform(X_test)
print(X_train)
```

Modeling With Logistic Regression

```python
logmodel = LogisticRegression(solver='lbfgs', max_iter=9000)
logmodel.fit(X_train, y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
 classification_report(y_test, predictions)
```

```python
confusion_matrix(y_test, predictions)
```

```python
accuracy_score(y_test, predictions)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python

print('The precision Score - we will diagnose yes heart disease correctly at this the percent:' +str(precision_score(y_test, predictions)))
print('The recall Score - we will diagnose yes heart disease correctly at this the percent:' +str(recall_score(y_test, predictions)))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict

scores = cross_val_score(logmodel, X_train, y_train, cv=15)
print('Cross-Validation Accuracy Scores',scores)
recall_scores = cross_val_score(logmodel, X_train, y_train, cv=15, scoring='recall')
print('Cross-Validation Recall Scores',recall_scores)
precision_scores = cross_val_score(logmodel, X_train, y_train, cv=15, scoring='precision')
print('Cross-Validation Precision Scores',precision_scores)
```

```python
scores = pd.Series(scores)
scores.min(), scores.mean(), scores.max()
```

```python
recall_scores = pd.Series(recall_scores)
recall_scores.min(), recall_scores.mean(), recall_scores.max()
```

```python
precision_scores = pd.Series(precision_scores)
precision_scores.min(), precision_scores.mean(), precision_scores.max()
```

Imagine there is a backlog for ECG analysis and an the decision is made to complete a preliminary analysis on partients before determing the next steps. If we have a patient with high probablity of heart disase based on quick data collection, we would act accordingly.Contrarily, if we had a patient with very low probablity of heart disease based on quick data collection, we may act differently. The intention here is to show how this tool could be useful in a fast moving & real life scenario.

```python
d = {}
d['MaxHR'] = float(input('What is the Max Heart Rate of the patient?'))
d['Age'] = float(input('How old is the patient?'))
d['Sex'] = input('What is the sex of the patient (M or F)?').casefold()
d['RestingBP'] = float(input('What is the resting BP reading of the patient?'))
d['Cholesterol'] = float(input('What is the cholesterol reading of the patient?'))
d['ExerciseAngina'] = input('Does the patient have exercise induced angina? (Y or N)').casefold()
d['FastingBS'] = float(input('What is the fasting blood sugar of the patient? [1: if FastingBS >120 mg/dl, 0: otherwise]'))
d['ChestPainType'] = input('Describe the chest pain type (use capitals): [TA: Typical Angina, ATA: Atypical Angina, NAP: Non-Anginal Pain, ASY: Asymptomatic]')

print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
print("Before getting an ECG evaluation on the patient, let's determine the current probability of heart disease")
```

```python
d
```

```python
for k,v in d.items():
  if k=='Sex':
    d.update({k: v.upper()})
  if k=='ExerciseAngina':
    d.update({k: v.upper()})
  if k=='ChestPainType':
    d.update({k: v.upper()})
```

```python
d
```

```python
unseen = pd.DataFrame(d.items(),index=d).T
unseen = unseen.drop([0])
```

```python
unseen1 = unseen.drop(columns=['Sex', 'ExerciseAngina', 'ChestPainType'])
unseen1
```
